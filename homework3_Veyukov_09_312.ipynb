1. Поиск ключей по значению: Напишите функцию, которая принимает словарь и значение, и возвращает список всех ключей, связанных с этим значением.

```python
def key_all_value(dct, value):
  return [i for i in dct if dct[i]==value]

my_dict = {'a': 1, 'b': 2, 'c': 1, 'd': 3}
print(key_all_value(my_dict, 1))
```

2. Добавление элементов: Напишите функцию, которая добавляет все элементы из списка в множество.

```python
def set_update(a, st):
  st.update(a)

a = [1, 2, 7, 4, 5]
st = {9, 2}
set_update(a, st)
print(st)
```

3. Разбиение кортежа: Напишите функцию, которая разбивает кортеж на две части: первую половину и вторую половину.



```python
def split_tuple(t):
    mid = len(t) // 2
    return t[:mid], t[mid:]

my_tuple = (1, 2, 3, 4, 5, 6, 7, 8)
first, second = split_tuple(my_tuple)

print(first)
print(second)
```

```python
a = [
  ['r', 'b', 'c'],
  ['b', 'c', 'a'],
  ['r', 'b', 'a'],
  ['c', 'a', 'b']
]
s = sorted(i for i in input())
k = 0
for i in s:
  for j in a:
    if i in j:
      k+=1
      break
print("+" if k==len(s) else "-")



```

```python
n = int(input())
k = int(input())
a = [i for i in range(1, n+1)]
f = open("demofile.txt", "w")
for i in a:
  f.write(str(i) + " ")
f.close()

f_from = open("demofile.txt", "r")
f_to = open("file.txt", "w")
for i in f_from:
  s = i.split()
  for j in s:
    if int(j) % k != 0:
      f_to.write(j + " ")
f_from.close()
f_to.close()
f = open("file.txt", "r")
print(f.readline())

```

```python
f = open("text.txt", "r")
d = {}
for s in f:
  s = s.lower().replace(".", "").split()
  for word in s:
    d[word] = d.get(word, 0) + 1
f.close()
print(max(d, key=d.get))
```
